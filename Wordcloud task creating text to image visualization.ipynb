```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Create a list of words
text1=("Sister HappyBirthday Sister HappyBirthday Sister HappyBirthday Happy PartyPoppers GodBlessYou BestWishes delicious friends FruitCake hangout joyfull bestday family MomDad Siblings Cousins relatives Partytime Restaurant invitations cooldrinks chips Gettogether chocolates balloons blissful")
```

```python
text1
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=500, height=500, margin=0).generate(text1)
```

```python
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
